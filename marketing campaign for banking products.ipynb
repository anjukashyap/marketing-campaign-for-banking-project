# train-test split

```python
# x is the input and y is the output
x=df.loc[:,df.columns!="Personal Loan"]
y=df[["Personal Loan"]]
```

```python
# importing library and classes
#dividing the data in 70:30  ration for training and testing repectively
from sklearn.model_selection import train_test_split
x_train, x_test ,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

# model training

```python
#classify function
from sklearn.model_selection import cross_val_score
def classify(model,x,y):
    x_train, y_train ,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
    model.fit(x_train,y_train)#training the model
    print("trainig Accuracy is",model.score(x_train,y_train)*100)
    print(" Testing Accuracy is ",model.score(x_test,y_test)*100)
    #cross validation - it is used for better visualization of model
    score=cross_val_score(model,x,y,cv=5)
    print("cross validation is ",np.mean(score)*100)
```

# Algorithm implementation

# logistic Regression

```python
#importing libraries and classes
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
classify(model,x,y)
```

# various other classification algorithms

# DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
classify(model,x,y)
```

# RandomForestClassifier  

```python
from sklearn.ensemble import RandomForestClassifier,ExtraTreesClassifier
model=RandomForestClassifier()
classify(model,x,y)
```

# ExtraTreesClassifier

```python
model=ExtraTreesClassifier()
classify(model,x,y)
```

# metrics for evaluating the model performance

```python
#y contains all the outputs and x contain all the inputs.we will test on the machine if it gives to expected output for the
#corresponding inputs
expected = y_test
predicted = model.predict(x_test)
```

```python
from sklearn import metrics
print(metrics.classification_report(expected,predicted))
```

# Compare various algorithm

```python
#import modules
from sklearn.datasets import make_classification
```

```python
# Defining learning classifiers
names = ["Logistic_Regression","Decision_Tree", "Extra_Trees", "Random_Forest"]
classifiers = [
    LogisticRegression(),
    DecisionTreeClassifier(),
    ExtraTreesClassifier(),
    RandomForestClassifier()
]
```

```python
#Build Model, Apply Model on Test Data & Record Accuracy Scores
scores = []
for name, clf in zip(names, classifiers):
    clf.fit(x_train, y_train)
    score = clf.score(x_test, y_test)
    scores.append(score)
```

```python
scores
```

```python
# Create data frame of model performance
df = pd.DataFrame()
df['name'] = names
df['score'] = scores
df
```
